# Geocoding the Bus Stops locations

```python
bus_stops=pd.read_csv('Bus_Stops.csv')
bus_stops.head()
```

```python
bus_positions=bus_stops.apply(lambda x: requests.get('https://geocode.search.hereapi.com/v1/geocode?q='+x['Street_One']+' '+x['Street_Two']+' san francisco',headers=headers3).json()['items'][0]['position'],axis=1)
```

```python
bus_Stops=bus_stops.join(pd.DataFrame(list(bus_positions)))
bus_Stops.head()
```

```python
bus_Stops.to_csv('Bus_Stops_with_positions.csv')
```
